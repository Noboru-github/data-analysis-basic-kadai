```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

fig, axes = plt.subplots(2, 3, tight_layout=True)

x0 = [0, 1, 2, 3]
y0 = [2, 3, -5, 6]
axes[0,0].plot(x0, y0)

x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(x1, y1, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))

x2 = np.random.normal(15, 5, 3000)
axes[0,2].hist(x2)

np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3, y3)

x4 = np.array([50, 20, 30])
x4_label = ["あさがお", "チューリップ", "さくら"]
axes[1,1].set_title("好きな花", loc="center")
axes[1,1].pie(x4, labels=x4_label)

x5 = np.array([1, 2, 3])
y5 = np.array([100, 80, 220])
x5_label = ["ビール", "焼酎", "日本酒"]
axes[1,2].set_title("酒類の消費量", loc="left")
axes[1,2].bar(x5, y5, tick_label=x5_label)


plt.show()
```
